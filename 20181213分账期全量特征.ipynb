```python
import pandas as pd
import numpy as np
training_data1_m1=pd.read_csv("contest_training_1.txt")
training_data2_m1=pd.read_csv("contest_training_2_1.csv")
training_data3_m1=pd.read_csv("contest_training_3_1.csv")
forecast_data1_m1=pd.read_csv("contest_forecast_1.txt")
forecast_data2_m1=pd.read_csv("contest_forecast_2_1.csv")
forecast_data3_m1=pd.read_csv("contest_forecast_3_1.csv")
training_data2_m1_1=pd.merge(training_data1_m1,training_data2_m1,on='user_id')
training_data_m1=pd.merge(training_data2_m1_1,training_data3_m1,on='user_id')
forecast_data2_m1_1=pd.merge(forecast_data1_m1,forecast_data2_m1,on='user_id')
forecast_data_m1=pd.merge(forecast_data2_m1_1,forecast_data3_m1,on='user_id')
```

```python
training_data1_m2=pd.read_csv("contest_training_1.txt")
training_data2_m2=pd.read_csv("contest_training_2_2.csv")
training_data3_m2=pd.read_csv("contest_training_3_2.csv")
forecast_data1_m2=pd.read_csv("contest_forecast_1.txt")
forecast_data2_m2=pd.read_csv("contest_forecast_2_2.csv")
forecast_data3_m2=pd.read_csv("contest_forecast_3_2.csv")
training_data2_m2_1=pd.merge(training_data1_m2,training_data2_m2,on='user_id')
training_data_m2=pd.merge(training_data2_m2_1,training_data3_m2,on='user_id')
forecast_data2_m2_1=pd.merge(forecast_data1_m2,forecast_data2_m2,on='user_id')
forecast_data_m2=pd.merge(forecast_data2_m2_1,forecast_data3_m2,on='user_id')
```

```python
training_data1_m3=pd.read_csv("contest_training_1.txt")
training_data2_m3=pd.read_csv("contest_training_2_3.csv")
training_data3_m3=pd.read_csv("contest_training_3_3.csv")
forecast_data1_m3=pd.read_csv("contest_forecast_1.txt")
forecast_data2_m3=pd.read_csv("contest_forecast_2_3.csv")
forecast_data3_m3=pd.read_csv("contest_forecast_3_3.csv")
training_data2_m3_1=pd.merge(training_data1_m3,training_data2_m3,on='user_id')
training_data_m3=pd.merge(training_data2_m3_1,training_data3_m3,on='user_id')
forecast_data2_m3_1=pd.merge(forecast_data1_m3,forecast_data2_m3,on='user_id')
forecast_data_m3=pd.merge(forecast_data2_m3_1,forecast_data3_m3,on='user_id')
```

数据集不大，可以使用corr()方法轻松计算出每对属性之间的标准相关系统（pearson）

```python
train1_m1=training_data_m1.drop("companycode",axis=1)
train2_m1=train1_m1.drop("product_id",axis=1)
train3_m1=train2_m1.drop("service_type",axis=1)
train4_m1=train3_m1.drop("area_id",axis=1)
train5_m1=train4_m1.drop("user_id",axis=1)
train6_m1=train5_m1.drop("month_id",axis=1)
train7_m1=train6_m1.drop("month_id_x",axis=1)
train8_m1=train7_m1.drop("month_id_y",axis=1).copy()
label_m1=training_data_m1['is_lost']
forecast1_m1=forecast_data_m1.drop("companycode",axis=1)
forecast2_m1=forecast1_m1.drop("product_id",axis=1)
forecast3_m1=forecast2_m1.drop("service_type",axis=1)
forecast4_m1=forecast3_m1.drop("area_id",axis=1)
forecast5_m1=forecast4_m1.drop("user_id",axis=1)
forecast6_m1=forecast5_m1.drop("month_id",axis=1)
forecast7_m1=forecast6_m1.drop("month_id_x",axis=1)
forecast8_m1=forecast7_m1.drop("month_id_y",axis=1).copy()
forecast_data_new_m1=forecast8_m1.drop("is_lost",axis=1).copy()
train9_m1=train8_m1.drop("is_lost",axis=1)
SX_m1=train9_m1.fillna(0)
y_SX_m1=forecast_data_new_m1.fillna(0)
```

```python
SX_m1.corr()
```

随机森林步骤：一先把相关性分析大于0.1的特征带入，二计算

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(100,oob_score=True)
rfc.fit(SX_m1,label_m1)
```

```python
from sklearn.metrics import classification_report
print(classification_report(label_m1,rfc.predict(SX_m1)))
```

```python
training_dftest_m1=training_data_m1[['user_id']].copy()
training_dftest_m1['probability']=rfc.predict(SX_m1)
training_result_new_m1=training_dftest_m1
training_result_new_m1.to_csv('train_m1.csv',index=False,header=False)
```

```python
forecast_dftest_m1=forecast_data_m1[['user_id']].copy()
forecast_dftest_m1['probability']=rfc.predict(y_SX_m1)
forecast_result_new_m1=forecast_dftest_m1
forecast_result_new_m1.to_csv('m1.csv',index=False,header=False)
```
